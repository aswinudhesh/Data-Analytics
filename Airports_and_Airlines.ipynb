```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_airports = pd.read_csv("/content/airports.dat.txt")
df_airports.head()
```

```python
 df_airports.columns = ["Airport_ID","Name","City","Country","IATA","ICAO","Latitude","Longitude","Altitude","Timezone","DST","Tz database timezone","Type","Source"]
 df_airports.head()
```

1)Top 20 highest altitude airports in the world.

```python
df_highest_altitude = df_airports.sort_values(by="Altitude", ascending=False)
df_highest_altitude['rank'] = df_highest_altitude['Altitude'].rank(ascending=False)
df_highest_altitude_sel = df_highest_altitude[['Airport_ID', 'Name', 'City', 'Country', 'Altitude', 'rank']]
df_top_20_highest_altitude = df_highest_altitude_sel.head(20)
df_top_20_highest_altitude.head(20)
```

```python
df_top_20_countries = df_airports.groupby(['Country']).size().reset_index(name='total_airports')
df_top_20_countries = df_top_20_countries.sort_values(by='total_airports', ascending=False)
df_top_20_countries['rank'] = df_top_20_countries['total_airports'].rank(ascending=False, method='min')
df_top_20_countries_sel = df_top_20_countries.head(20)
print(df_top_20_countries_sel)
```
